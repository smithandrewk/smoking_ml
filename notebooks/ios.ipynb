```python
import pandas as pd
df = pd.read_csv(f'../SensorLogFiles_my_iOS_device_230601_09-48-47/2023-05-31_20_23_09_Apple Watch.csv')
df = df.rename({'loggingTime(txt)':'t','accelerometerTimestamp_sinceReboot(s)':'t_acc','accelerometerAccelerationX(G)':'acc_x','accelerometerAccelerationY(G)':'acc_y','accelerometerAccelerationZ(G)':'acc_z','batteryLevel(R)':'battery'},axis=1)
df = df.drop('batteryState(N)',axis=1)
df['t'] = df['t'].astype('datetime64')

df
```

```python
import plotly.express as px
fig = px.line(data_frame=df,y=['acc_x','acc_y','acc_z'])
fig.show(renderer='browser')
```
